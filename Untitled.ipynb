```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rupankarmajumdar/crop-pests-dataset")

print("Path to dataset files:", path)
```

```python
import cv2
import os

image_paths = []
label_paths = []
train_folder = r"C:\Users\dell\.cache\kagglehub\datasets\rupankarmajumdar\crop-pests-dataset\versions\2\train"
train_images_folder = os.path.join(train_folder, "images")
train_labels_folder = os.path.join(train_folder, "labels")
    
for name in os.listdir(train_images_folder):
    full_path = os.path.join(train_images_folder, name)
    image_paths.append(full_path)

for name in os.listdir(train_labels_folder):     
    full_path = os.path.join(train_labels_folder, name)
    label_paths.append(full_path)

image_paths = sorted(image_paths)
label_paths = sorted(label_paths)
```

```python
import numpy

x = []
y = []
for img_path, label_path in zip(image_paths, label_paths):
    img = cv2.imread(img_path)
    h, w, _ = img.shape
    with open(label_path, "r") as f:
        lines = f.readlines()
    for line in lines:
        cls, x_c, y_c, bw, bh = map(float, line.strip().split())
        x1 = int((x_c - bw/2) * w)
        y1 = int((y_c - bh/2) * h)
        x2 = int((x_c + bw/2) * w)
        y2 = int((y_c + bh/2) * h)
        roi = img[y1:y2, x1:x2]
        if roi.size == 0: 
            continue
        roi = cv2.resize(roi, (64, 64))
        hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0, 1, 2], None, [8, 8, 8],
                            [0, 180, 0, 256, 0, 256])
        hist = cv2.normalize(hist, hist).flatten()
        x.append(hist)
        y.append(int(cls)) 
    del img
```

```python
import cv2
import numpy as np

from skimage.feature import local_binary_pattern
from sklearn.preprocessing import StandardScaler

# ========== LBP 函数 ==========
def extract_LBP(gray):
    lbp = local_binary_pattern(gray, P=8, R=1, method="uniform")
    hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 59+1), range=(0,59))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)
    return hist

# ========== 初始化 ==========
X = []
y = []

hog = cv2.HOGDescriptor(
    _winSize=(64, 64),
    _blockSize=(16, 16),
    _blockStride=(8, 8),
    _cellSize=(8, 8),
    _nbins=9
)

# ==========  特征提取  ==========
for img_path, label_path in zip(image_paths, label_paths):

    img = cv2.imread(img_path)
    if img is None:
        continue

    h_img, w_img, _ = img.shape

    with open(label_path, "r") as f:
        lines = f.readlines()

    for line in lines:
        cls, x_c, y_c, bw, bh = map(float, line.strip().split())

        x1 = int((x_c - bw/2) * w_img)
        y1 = int((y_c - bh/2) * h_img)
        x2 = int((x_c + bw/2) * w_img)
        y2 = int((y_c + bh/2) * h_img)

        roi = img[y1:y2, x1:x2]

        # ROI 过滤
        if roi.size == 0 or roi.shape[0] < 20 or roi.shape[1] < 20:
            continue

        roi = cv2.resize(roi, (64, 64))

        # --- HSV ---
        hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0,1,2], None, [8,8,8],
                            [0,180,0,256,0,256])
        hist = cv2.normalize(hist, hist).flatten()

        # --- Hu ---
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        hu = cv2.HuMoments(cv2.moments(gray)).flatten()

        # --- LBP ---
        lbp_feat = extract_LBP(gray)

        # --- HOG ---
        h = hog.compute(roi).flatten()

        # --- 合并特征 ---
        feature = np.hstack([hist, hu, lbp_feat, h])

        X.append(feature)
        y.append(int(cls))

print("🎉 特征提取完成")
X = np.array(X)
y = np.array(y)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

model = make_pipeline(
    StandardScaler(),
    SVC(kernel='rbf', C=10, gamma='scale', class_weight='balanced')
)

model.fit(X, y)
```

```python
import cv2
import os

test_image_paths = []
test_label_paths = []
test_folder = r"C:\Users\dell\.cache\kagglehub\datasets\rupankarmajumdar\crop-pests-dataset\versions\2\test"
test_images_folder = os.path.join(test_folder, "images")
test_labels_folder = os.path.join(test_folder, "labels")
    
for name in os.listdir(test_images_folder):
    full_path = os.path.join(test_images_folder, name)
    test_image_paths.append(full_path)

for name in os.listdir(test_labels_folder):     
    full_path = os.path.join(test_labels_folder, name)
    test_label_paths.append(full_path)

test_image_paths = sorted(test_image_paths)
test_label_paths = sorted(test_label_paths)
```

```python
X_test = []
y_test = []

hog = cv2.HOGDescriptor(
    _winSize=(64, 64),
    _blockSize=(16, 16),
    _blockStride=(8, 8),
    _cellSize=(8, 8),
    _nbins=9
)

for img_path, label_path in zip(test_image_paths, test_label_paths):

    img = cv2.imread(img_path)
    if img is None:
        continue

    h_img, w_img, _ = img.shape

    with open(label_path, "r") as f:
        lines = f.readlines()

    for line in lines:
        cls, x_c, y_c, bw, bh = map(float, line.strip().split())

        x1 = int((x_c - bw/2) * w_img)
        y1 = int((y_c - bh/2) * h_img)
        x2 = int((x_c + bw/2) * w_img)
        y2 = int((y_c + bh/2) * h_img)

        roi = img[y1:y2, x1:x2]

        if roi.size == 0 or roi.shape[0] < 20 or roi.shape[1] < 20:
            continue

        roi = cv2.resize(roi, (64, 64))

        # --- HSV ---
        hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)
        hist = cv2.calcHist([hsv], [0,1,2], None,
                            [8,8,8],
                            [0,180,0,256,0,256])
        hist = cv2.normalize(hist, hist).flatten()

        # --- Hu 矩 ---
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        hu = cv2.HuMoments(cv2.moments(gray)).flatten()

        # --- LBP ---
        lbp_feat = extract_LBP(gray)

        # --- HOG ---
        h = hog.compute(roi).flatten()

        # --- 合并特征 ---
        feature = np.hstack([hist, hu, lbp_feat, h])

        X_test.append(feature)
        y_test.append(int(cls))

print("🎉 Test 特征提取完成")

from sklearn.metrics import accuracy_score, classification_report

X_test = np.array(X_test)
y_test = np.array(y_test)

# pipeline 会自动做 StandardScaler().transform()
y_pred = model.predict(X_test)

print("Test 准确率:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
